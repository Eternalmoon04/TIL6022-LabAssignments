```python
import pandas as pd
import plotly.express as px
```

### data processing

```python
# read csv files
file_path = 'flightlist'
file_path_addition = ['_20191001_20191031','_20191101_20191130','_20191201_20191231','_20200101_20200131','_20200201_20200229','_20200301_20200331',
'_20200401_20200430','_20200501_20200531','_20200601_20200630','_20200701_20200731','_20200801_20200831','_20200901_20200930','_20201001_20201031',
'_20201101_20201130','_20201201_20201231','_20210101_20210131','_20210201_20210228','_20210301_20210331','_20210401_20210430','_20210501_20210530',
'_20210601_20210630','_20210701_20210731','_20210801_20210831','_20210901_20210930','_20211001_20211031']
dfall = pd.DataFrame()
for i in file_path_addition:
    df = pd.read_csv(file_path + i +'.csv')
    dfall = dfall.append(df)

#get useful columns
dfall = pd.DataFrame(dfall, columns=['callsign','origin','destination','day'])

#drop incomplete OD values and flights have the same OD
dfall.dropna(axis=0, how='any', inplace=True)
dfall = dfall.drop(index=dfall[dfall['origin']==dfall['destination']].index)

#output to another csv file
dfall.to_csv('flightdata.csv', index=False)
```

### zones

#### Mid-East

```python
fd_mideast_map = fd[( #Middle East
            fd['origin'] == 'OMDB') |
           (fd['origin'] == 'OTHH') ]
fd_mideast_group = fd_mideast_map.groupby(['origin', 'day'], as_index=False).count()
fig = px.line(fd_mideast_group , x="day", y="callsign", color="origin")
fig.update_xaxes(title={'text': 'flight of middle east'})
fig.update_yaxes(title={'text': 'Number of flights'})
fig.show()
```

#### Europe

```python
fd_europe_map = fd[ #europe
           (fd['origin'] == 'EHAM') |
           (fd['origin'] == 'EDDF') |
           (fd['origin'] == 'LFPG') |
           (fd['origin'] == 'EGLL') |
           (fd['origin'] == 'LEMD') ]
fd_europe_group = fd_europe_map.groupby(['origin', 'day'], as_index=False).count()
fig = px.line(fd_europe_group , x="day", y="callsign", color="origin")
fig.update_xaxes(title={'text': 'flight of europe'})
fig.update_yaxes(title={'text': 'Number of flights'})
fig.show()
```
